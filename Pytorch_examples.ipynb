# linear Regression

```
%matplotlib inline
```

```
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision

device = torch.device('cuda:0' if torch.cuda.is_available() else "cpu")
```

```
from pprint import pprint

import matplotlib.pyplot as plt
import numpy as np
from IPython.core.debugger import set_trace
```

```
from sklearn.datasets import make_regression

n_features = 1
n_samples = 10000

x, y = make_regression(
    n_samples=n_samples,
    n_features=n_features,
    noise=20,
    random_state=42
)

fix, ax = plt.subplots()
ax.plot(x, y, '.')
```

```
x = torch.from_numpy(x).float()
y = torch.from_numpy(y.reshape((n_samples, n_features))).float()
```

```
x.size(), y.size()
```

```
x[0], y[0]
```

```
class LinReg(nn.Module):
  
  def __init__(self, input_dim):
    super().__init__()
    self.beta = nn.Linear(input_dim, 1)
  
  def forward(self, x):
    return self.beta(x)
```

```
model = LinReg(n_features).to(device)
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.001)

x, y = x.to(device), y.to(device)
```

```
epochs = True
epoch = 0
val_freq = 1
while epochs:
  val_freq += 1
  epoch += 1
  least_loss = 0.001
  model.train()
  y_ = model(x)
  loss = loss_fn(y_, y)
  loss.backward()
  optimizer.step()
  print('-- Epoch : {} -- Loss : {}'.format(epoch, loss.item()))
  if val_freq % 100 == 0:
    model.eval()
    with torch.no_grad():
      y_ = model(x)
    fig, ax = plt.subplots()
    ax.plot(x.cpu().numpy(), y_.cpu().numpy(), '.', label="pred")
    ax.plot(x.cpu().numpy(), y.cpu().numpy(), '.', label="data")
    ax.set_title(f"MSE: {loss.item():0.1f}")
    ax.legend();
    plt.show()
  if loss.item() < least_loss:
    epochs = False
  if epoch > 1000:
    break
```

```
def get_trainable(model_params):
  return (p for p in model_params if p.requires_grad)

def get_frozen(model_params):
  return (p for p in model_params if not p.requires_grad)

def all_trainable(model_params):
  return all(p.requires_grad for p in model_params)

def all_frozen(model_params):
  return all(not p.requires_grad for p in model_params)

def freeze_all(model_params):
  for param in model_params:
    param.requires_grad = False
```

```
from torchvision import transforms
```

```
_image_size = 224
_mean = [0.485, 0.456, 0.406]
_std = [0.229, 0.224, 0.225]

train_trans = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomCrop(_image_size),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(.3, .3, .3),
    transforms.ToTensor(),
    transforms.Normalize(_mean, _std)
])

val_trans = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(_image_size),
    transforms.ToTensor(),
    transforms.Normalize(_mean, _std)
])
```

```
import os
import zipfile

from torchvision.datasets.folder import ImageFolder, default_loader
from torchvision.datasets.utils import download_url, check_integrity
```

```
train_ds = ImageFolder('dogscats/train', transform=train_trans)
test_ds = ImageFolder('dogscats/valid', transform=val_trans)
```

```
from torch.utils.data import DataLoader
```

```
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True, num_workers=4)
val_loader = DataLoader(test_ds, batch_size=32, shuffle=True, num_workers=4)
```

```
from torchvision import models
```

```
n_classes = 2
```

```
def get_model(n_classes=2):
  model = models.resnet18(pretrained=True)
  freeze_all(model.parameters())
  model.fc = nn.Linear(512, n_classes)
  model = model.to(device)
  return model

resnet_model = get_model()
```

```
import torchsummary

torchsummary.summary(resnet_model, (3, 224, 224))
```

```
criterion = nn.CrossEntropyLoss()
```

```
optimizer = torch.optim.Adam(get_trainable(resnet_model.parameters()), lr=1e-3)
```

```
n_epochs = 2

for epoch in range(n_epochs):
  print('Epoch {} /{}'.format(epoch +1, n_epochs))
  
  resnet_model.train()
  
  running_loss, correct = 0.0, 0
  for x, y in train_loader:
    x, y = x.to(device), y.to(device)
    
    optimizer.zero_grad()
    y_ = resnet_model(x)
    loss = criterion(y_, y)
    loss.backward()
    optimizer.step()
    
    # statistics
    print(f"batch loss: {loss.item():0.3f}")
    _, y_label_ = torch.max(y_, 1)
    correct += (y_label_ == y).sum().item()
    running_loss += loss.item() * x.shape[0]
  
  print(f" Train Loss: {running_loss/len(train_loader.dataset)}")
  print(f" Train Acc : {correct / len(train_loader.dataset)}")
  
  resnet_model.eval()
  
  running_loss, correct = 0.0, 0
  with torch.no_grad():
    for x, y in val_loader:
      x, y = x.to(device), y.to(device)
      y_ = resnet_model(x)
      
      # statistics
      _, y_label_ = torch.max(y_, 1)
      correct += (y_label_ == y).sum().item()
      loss = criterion(y_, y)
      running_loss += loss.item() * x.shape[0]
  print(f"Valid Loss: {running_loss/len(val_loader.dataset)}")
  print(f"Valid Acc: {correct/len(val_loader.dataset)}")
  print()
```
